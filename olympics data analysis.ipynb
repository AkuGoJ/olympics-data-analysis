```python
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline
```

### Load Dataset 

```python
athletes = pd.read_csv('athlete_events.csv')
regions = pd.read_csv('noc_regions.csv')
```

```python
athletes.head()
```

```python
regions.head()
```

### Join The DataFrame 

```python
athletes_df = athletes.merge(regions, how='left', on='NOC')
athletes_df.head()
```

```python
athletes_df.shape
```

### Column name consistent 

```python
athletes_df.rename(columns={'region':'Region', 'notes':'Notes'}, inplace=True)
```

```python
athletes_df.head()
```

### Info Method

### Top Countries participating 

```python
top_10_countries = athletes_df.Team.value_counts().sort_values(ascending=False).head(10)
```

```python
top_10_countries
```

### ploting for the top 10 countries 

```python
plt.figure(figsize=(12,6))
plt.title('Top 10 Countries participation')
sns.barplot(x=top_10_countries.index,y=top_10_countries, palette='Set2')
plt.xticks(rotation=-30)
plt.show()
```

### Age Distribution of the participants 

```python
plt.figure(figsize=(12,6))
plt.title('Age distribution of the athletes')
plt.xlabel('Age')
plt.ylabel('Number of Participants')
plt.hist(athletes_df.Age,bins=np.arange(10,80,2),color='orange',edgecolor='black')
plt.show()
```

### Total number of female athletes in each Olympics 

```python
female_participants = athletes_df[(athletes_df.Sex=='F') & (athletes_df.Season=='Summer')][['Sex','Year']]
female_participants = female_participants.groupby('Year').count().reset_index()
female_participants.head()
```

### Another Method 

```python
womenolympics = athletes_df[(athletes_df.Sex == 'F') &(athletes_df.Season == 'Summer')]
```

```python
sns.set(style='darkgrid')
plt.figure(figsize=(20,10))
sns.countplot(x='Year', data=womenolympics, palette='Spectral')
plt.title('Women Participations')
plt.show()
```

### Female Athlets over the times


```python
plt.figure(figsize=(16,8))
plt.title('Female Athlets over the times')
part = womenolympics.groupby('Year')['Sex'].value_counts()
part.loc[:,'F'].plot()
plt.show()
```

### Gold Medal Athlets

```python
goldmedal= athletes_df[(athletes_df.Medal == 'Gold')]
```

```python
goldmedal.head()
```

### Take only the values that are diffrent from NaN 

```python
goldmedal = goldmedal[np.isfinite(goldmedal['Age'])]
```

### Gold Beyond 60 

```python
goldmedal['ID'][goldmedal['Age']>60].count()
```

```python
sporting_event = goldmedal['Sport'][goldmedal['Age']>60]
sporting_event
```

### plot for sporting Events 

```python
plt.figure(figsize=(12,6))
plt.tight_layout()
sns.countplot(sporting_event)
plt.title(' gold medals for athlets over 60 years')
plt.show()
```

### Gold medals from each country

```python
goldmedal.Region.value_counts().reset_index(name='Medal').head(5)
```

### Total Gold Medals 

```python
totalGoldMedals = goldmedal.Region.value_counts().reset_index(name='Medal').head(6)
```

```python
totalGoldMedals
```

### making graph on Total number of gold medals 

```python
g = sns.catplot(x='index',y='Medal',data=totalGoldMedals, height=5,kind='bar',palette='rocket')
g.despine(left=True)
g.set_xlabels('Top 5 Countries')
g.set_ylabels('Number of medals')
plt.title('Gold medals par country')
plt.show()
```
